```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (18, 7))
plt.figure(figsize = (8, 6))
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')
axes[0].set_title('Normal Model Confusion Matrix')
sns.heatmap([[354, 680],
            [74, 335]], annot = True, cmap = 'Blues', fmt = 'd', xticklabels = ['Predicted Negative', 'Predicted Positive'], yticklabels = ['True Negative', 'True Positive'], ax = axes[0])

axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')
axes[1].set_title('Fact-Checking Pipeline Confusion Matrix')
sns.heatmap([[967, 67],
            [404, 5]], annot = True, cmap = 'Blues', fmt = 'd', xticklabels = ['Predicted Negative', 'Predicted Positive'], yticklabels = ['True Negative', 'True Positive'], ax = axes[1])

plt.show()
```
